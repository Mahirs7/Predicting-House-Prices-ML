```python
df.drop(['index', 'full_link'], axis=1, inplace=True)
```

```python
df.drop(['title', 'bedrooms', 'description', 'mls', 'type', 'full_address', 'lat', 'long', 'city_district', 'district_code'], axis=1, inplace=True)
df.drop('final_price', axis=1, inplace=True)
```

```python
df
```

```python
df.dropna(inplace=True)
```

```python
df2 = df[['bathrooms', 'sqft', 'parking', 'mean_district_income', 'bedrooms_ag', 'bedrooms_bg', 'final_price_log']]
df2
```

```python
from sklearn.model_selection import train_test_split

X = df2[['bathrooms', 'sqft', 'parking', 'mean_district_income', 'bedrooms_ag', 'bedrooms_bg']].values
y = df2.iloc[:, -1].values

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
model = RandomForestRegressor()
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import r2_score

r2_score(y_test, y_pred)
```
